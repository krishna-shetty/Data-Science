```python
from selenium import webdriver
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import time
import requests
```

```python
driver = webdriver.Chrome(executable_path = 'C:/Users/pc/Downloads/chromedriver_win32 (1)/chromedriver.exe')
```

```python
reviews_url = 'https://www.ign.com/reviews/games'
driver.get(reviews_url)
ScrollNumber = 30
for i in range(1,ScrollNumber):
    driver.execute_script("window.scrollTo(1, 50000)")
    time.sleep(1)
content = driver.page_source
```

```python
driver.execute_script('''window.open("http://www.ign.com/?setccpref=US","_blank");''')
soup = BeautifulSoup(content)
game_article_tags = soup.find_all('a', class_ = 'item-body')
print(len(game_article_tags))
```

```python
articles_base_url = 'https://www.ign.com'
titles = []
urls = []

for i in range(len(game_article_tags)):
    titles.append(game_article_tags[i]['aria-label'])
    urls.append(articles_base_url + game_article_tags[i]['href']) 

review_links_dict = {'title': titles, 'url': urls}
review_links_df = pd.DataFrame(review_links_dict)
review_links_df
```

```python
review_links_df.shape
```

```python
titles = []
developers = []
publishers = []
release_dates = []
platforms = []
genres = []
scores = []
reviewed_by = []

for url in review_links_df['url']:
    driver.get(url)
    content = driver.page_source
    soup_games = BeautifulSoup(content)
    title_tags = soup_games.find_all('h1', class_ = "jsx-3409712203")
    if len(title_tags) == 0:
        titles.append(np.nan)
    else:
        titles.append(title_tags[0].text.replace('Review', ''))
    
    
    developers_tags = soup_games.find_all('div', class_ = "jsx-3421308495 developers-info line-clamp")
    if len(developers_tags) == 0:
        developers.append(np.nan)
    else:
        developers.append(developers_tags[0].text[10:])
    
    
    publishers_tags = soup_games.find_all('div', class_ = "jsx-3421308495 publishers-info line-clamp")
    if len(publishers_tags) == 0:
        publishers.append(np.nan)
    else:
        publishers.append(publishers_tags[0].text[10:])
    
    
    release_date_tags = soup_games.find_all('div', class_ = "jsx-3421308495 release-date-info line-clamp")
    if len(release_date_tags) == 0:
        release_dates.append(np.nan)
    else:
        release_dates.append(release_date_tags[0].text[12:])
    
    
    platforms_tags = soup_games.find_all('div', class_ = "jsx-3421308495 platforms-info line-clamp")
    if len(platforms_tags) == 0:
        platforms.append(np.nan)
    else:
        platforms.append(platforms_tags[0].text[9:])
    
    
    genres_tags = soup_games.find_all('div', class_ = "jsx-3421308495 genres-info line-clamp")
    if len(genres_tags) == 0:
        genres.append(np.nan)
    else:
        genres.append(genres_tags[0].text[6:])
    
    
    score_tags = soup_games.find_all('span', class_ = "jsx-3567619958 hexagon-content")
    if len(score_tags) == 0:
        scores.append(np.nan)
    else:
        scores.append(float(score_tags[0].text))
    
    
    by_tags = soup_games.find_all('section', class_ = 'jsx-2034901901 author-names')
    if len(by_tags) == 0:
        reviewed_by.append(np.nan)
    else:
        reviewed_by.append(by_tags[0].text[3:])
```

```python
reviews_dict = {'title': titles, 'developers': developers, 'publishers': publishers, 'release_date': release_dates, 'platforms': platforms, 'genres': genres, 'score': scores, 'reviewed_by': reviewed_by}
reviews_df = pd.DataFrame(reviews_dict)
reviews_df
```

```python
reviews_df['developers'].fillna('-', inplace = True)
developers = []
for x in reviews_df['developers']:
    if x == '-':
        lst = np.nan
    else:
        lst = x.split(',') 
    developers.append(lst)
reviews_df['developers'] = developers

reviews_df['platforms'].fillna('-', inplace = True)
platforms = []
for x in reviews_df['platforms']:
    if x == '-':
        lst = np.nan
    else:
        lst = x.split(',')
    platforms.append(lst)
reviews_df['platforms'] = platforms

reviews_df['genres'].fillna('-', inplace = True)
genres = []
for x in reviews_df['genres']:
    if x == '-':
        lst = np.nan
    else:
        lst = x.split(',')
    genres.append(lst)
reviews_df['genres'] = genres

reviews_df
```

```python
reviews_df.to_csv('data/reviews.csv', index = False)
```
